```python
def anomalie(bullo):
  df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/mmsa/Test2/2B/reso/%s' % bullo).drop(columns=['Unnamed: 0']).replace(np.nan, 0)
  anomalie = pd.DataFrame(columns = ["NameUser", "NameActivity", "PredictedActivity"])
  index_anom = 0
  for row in range(0,len(df)-1):
    max = 0
    max_col = 0
    for col in df.columns[1:]:
      if(df.at[row,col]>=max):
        max = df.at[row,col]
        max_col = col
    anomalie.at[index_anom,anomalie.columns[0]] = bullo.split('_')[3]
    anomalie.at[index_anom,anomalie.columns[1]] = df.at[row, "NameActivity"]
    anomalie.at[index_anom,anomalie.columns[2]] = max_col
    index_anom +=1

  return anomalie
```

per ogni file si vede qual'è l'attività prevalente durante ogni schermata, selezionata viene inserita in un nuovo csv che quindi per ogni schermata riferisce l'attività svolta

```python
for file in os.listdir('/content/gdrive/MyDrive/Colab Notebooks/mmsa/Test2/2B/reso'):
  if(not os.path.exists('/content/gdrive/MyDrive/Colab Notebooks/Test2/2B/2B_tabella/%s' % file)):
    df=anomalie(file)
    print('fatto')
    df.to_csv('/content/gdrive/MyDrive/Colab Notebooks/mmsa/Test2/2B/2B_tabella/%s' % file)
```
