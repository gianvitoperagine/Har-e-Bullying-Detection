```python
arr = []
print(os.listdir(os.getcwd()))

for file in os.listdir(os.getcwd()):
    if file in no:
        print('\n')
        print(file)
        print('Questo non verrà estratto!\n')
    else:
        print(file)
        w1, w2 = str(file).split(".")
        print(w1, w2)
        path = '/content/drive/MyDrive/Colab Notebooks/mmsa/Test1/csv_estratti'

        if not os.path.exists(path + '/__%s_ACCELEROMETER.csv' % w1):

            if file[7].isalpha():
                df = pd.read_csv(file, delimiter=' - ', skipinitialspace=True, error_bad_lines=False, header=None, names=['DATETIME', 'ACTIVITY', 'DATA', 'NUL']).drop(['NUL'], axis=1)
            else:
                df = pd.read_csv(file, delimiter=';', skipinitialspace=True, header=None, error_bad_lines=False, names=['DATETIME', 'ACTIVITY', 'DATA', 'ID', 'NUL']).drop(['ID', 'NUL'], axis=1)

            word = 'ACCELEROMETER' if file[7].isalpha() else 'accelerometer'
            df['mask_df1'] = np.where(df['ACTIVITY'].str.contains(word, na=False), True, False)

            print(df[df['mask_df1'] == True])

            acc_activities = df[df['mask_df1'] == True]
            acc_activities.reset_index(inplace=True, drop=False)
            arr.append(len(acc_activities['ACTIVITY']))
            print(len(acc_activities['ACTIVITY']))

            df1 = pd.DataFrame(columns=['NameActivity', 'X', 'Y', 'Z'])
            i = 0
            for item in acc_activities['ACTIVITY']:
                if file[7].isalpha():
                    f1, f2, f3 = [x[1:] for x in acc_activities['DATA'][i].split(',')]
                else:
                    f1, f2, f3 = [x[2:] for x in acc_activities['DATA'][i].split('_')]
                print(f1, f2, f3)

                if len(acc_activities['ACTIVITY'][i].split('ACCELEROMETER_')) == 1:
                    activity = acc_activities['ACTIVITY'][i].split('ACCELEROMETER_')
                else:
                    temp, activity = acc_activities['ACTIVITY'][i].split('ACCELEROMETER_')
                df1.loc[i] = [activity, f1, f2, f3]

                i += 1

            if not os.path.isdir(path):
                os.mkdir(path)

            if not os.path.exists(path + '/__%s_ACCELEROMETER.csv' % w1):
                print('Creato perché inesistente')
                df1.to_csv(path + '/__%s_ACCELEROMETER.csv' % w1)
            else:
                print('Aggiornato perché già esistente')
                df1.to_csv(path + '/__%s_ACCELEROMETER.csv' % w1, mode='a', header=False)
        else:
            print('File già esistente!')

print(np.max(arr))
print(len(arr))
max_val = np.max(arr)
```
